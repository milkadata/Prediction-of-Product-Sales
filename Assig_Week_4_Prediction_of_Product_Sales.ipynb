### Missing data

```python
df.groupby(['Item_Type'])['Item_Weight'].mean()
```

```python
df.groupby(['Item_Fat_Content'])['Item_Weight'].mean()
```

```python
#Replacing weight with the average because other items are not that affected by it
median_weight = df['Item_Weight'].median()
df['Item_Weight'].fillna(median_weight, inplace=True)
df.isna().sum()
```

```python
#Chose to replace them with because Size is a category & 2,410 is more than 20% missing so don't want to give it a "most common value". That might give us false results 
df['Outlet_Size'].fillna('Missing', inplace=True)
df.isna().sum()
```

```python
df.nunique()
```

```python
df['Item_Fat_Content'].value_counts()
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('LF', 'Low Fat')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('reg', 'Reg')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('low fat', 'Low Fat')
df['Item_Fat_Content'].value_counts()
```

```python
# COMBINED style
with plt.style.context(['ggplot','seaborn-v0_8-dark-palette', 'fivethirtyeight']):
    ax = df['Item_Weight'].hist(bins = 50, edgecolor='white')
    ax.tick_params(axis='x', rotation = 45)
    ax.ticklabel_format(style='plain')
    ax.set_title('Item Weight')
```

```python
def my_scatterplot(df, x='Item_Outlet_Sales', y='Item_Weight'):
    fig, ax = plt.subplots()
    ax.scatter(df[x], df[y], edgecolor='white')
    ax.set(title=f'{x} vs. {y}', ylabel=y, xlabel=x);
    return fig, ax
```

```python
fig, ax = my_scatterplot(df)
```

```python
year_means = df.groupby('Outlet_Establishment_Year').mean()[['Item_MRP','Item_Weight','Item_Outlet_Sales']]
year_means.head()
```

```python
year = year_means.index
year
```

```python
mean_sales = year_means['Item_Outlet_Sales'].values
mean_sales
```

```python
fig, ax = plt.subplots()
ax.plot(year, mean_sales,c = 'y', marker = '*', markersize = 5);
ax.set(xlabel='Outlet_Establishment_Year', ylabel='Item_Outlet_Sales');
ax.set_title('Sales by Year of Outlet Establishment Year')
```
